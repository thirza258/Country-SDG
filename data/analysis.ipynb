```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter Afghanistan data
afghanistan = dataset[dataset['country'] == 'Afghanistan']

# Get all goal score columns (excluding 'country', 'year', etc.)
goal_columns = [col for col in dataset.columns if 'goal_' in col]

# Set up subplots
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(15, 18))  # Adjust layout for 17 goals
axes = axes.flatten()  # Flatten to easily iterate over

# Plot each goal in a separate subplot
for i, goal in enumerate(goal_columns):
    axes[i].plot(afghanistan['year'], afghanistan[goal], marker='o', linestyle='-', label=goal.replace('_', ' ').title(), color='b')
    
    # Labels and styling
    axes[i].set_title(goal.replace('_', ' ').title(), fontsize=12, fontweight='bold')
    axes[i].set_xlabel("Year", fontsize=10)
    axes[i].set_ylabel("Score", fontsize=10)
    axes[i].grid(True, linestyle="--", alpha=0.7)
    axes[i].set_xticks(afghanistan['year'])  # Ensure all years are shown
    axes[i].set_xticklabels(afghanistan['year'], rotation=45)  # Rotate for better readability

# Adjust layout
plt.tight_layout()
plt.show()
```
